Population vs. Player Count

```python
# Merge the player counts with state population data
state_analysis = state_player_counts.merge(places_population, on='state', how='left')
state_analysis = state_analysis.merge(states_population, on='state', how='left')
```

```python
state_analysis.head()
```

```python
# Calculate players per capita (for 2016 and 2017)
state_analysis['players_per_capita_2016'] = state_analysis['player_count'] / state_analysis['pop_2016']
state_analysis['players_per_capita_2017'] = state_analysis['player_count'] / state_analysis['pop_2017']
```

```python
# Display the analysis
state_analysis.info()
```

```python
# Visualize the data
# Bar plot of players per state
plt.figure(figsize=(14, 8))
sns.barplot(x='state', y='player_count', data=state_analysis)
plt.xticks(rotation=90)
plt.title('Number of Players per State')
plt.xlabel('State')
plt.ylabel('Player Count')
plt.show()
```

```python
# Scatter plot of players per capita
plt.figure(figsize=(14, 8))
sns.scatterplot(x='pop_2017', y='players_per_capita_2017', data=state_analysis)
plt.title('Players per Capita vs Population (2017)')
plt.xlabel('Population (2017)')
plt.ylabel('Players per Capita (2017)')
plt.show()
```

```python
# Top birthplaces by city
top_cities = player_data['player_birthplace_(combo)'].value_counts().head(10)
print("Top Cities by Player Birthplaces:\n", top_cities)

# Top birthplaces by state
player_data['State'] = player_data['player_birthplace_(combo)']
top_states = player_data['State']
print("Top States by Player Birthplaces:\n", top_states)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot for Top Cities by Player Birthplaces
plt.figure(figsize=(10, 6))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='viridis')
plt.title('Top 10 Cities by Super Bowl Player Birthplaces', fontsize=16)
plt.xlabel('Number of Players', fontsize=12)
plt.ylabel('City', fontsize=12)
plt.show()

# Plot for Top States by Player Birthplaces
plt.figure(figsize=(10, 6))
sns.barplot(x=top_states.values, y=top_states.index, palette='magma')
plt.title('Top 10 States by Super Bowl Player Birthplaces', fontsize=16)
plt.xlabel('Number of Players', fontsize=12)
plt.ylabel('State', fontsize=12)
plt.show()
```
